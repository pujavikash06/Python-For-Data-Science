#Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

#Data

```python
iris_data=pd.read_csv('../data/iris.csv')
print iris_data['SepalLengthCm'].max(),iris_data['SepalLengthCm'].min()
print iris_data['SepalWidthCm'].max(),iris_data['SepalWidthCm'].min()
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
fig=iris_data.plot(kind='Scatter',figsize=(12,5),x='SepalLengthCm',y='SepalWidthCm',color='y')
fig.set_ylim(2,5)
fig.set_xlim(4,8)
fig.set_title('SepalLength Vs SepalWidth')

#iris_data.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
```

```python
#Using Seaborn Jointplot to show bivariate scatterplots and univariate histograms
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris_data, size=5)
```

```python
#WE HAVE COLOURED DIFFERENT SPECIES
sns.FacetGrid(iris_data, hue="Species", size=10) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
#A look at different individual features through boxplots
sns.boxplot(x="Species", y="PetalLengthCm", data=iris_data)
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_data)
```

```python
#Another useful plot is the pairplotting 
sns.pairplot(iris_data.drop('Id',axis=1),hue='Species',size=5)
#We can see that iris-setosa is separated from the other species
```

```python
sns.pairplot(iris_data.drop('Id',axis=1),hue='Species',size=5,diag_kind='kde')
```

```python
#Andrews Curves
from pandas.tools.plotting import andrews_curves
andrews_curves(iris_data.drop("Id", axis=1), "Species")
```
